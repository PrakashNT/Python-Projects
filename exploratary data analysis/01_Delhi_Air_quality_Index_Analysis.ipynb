Calculate the air quality index from the available concentration metrials

```python
#define air quality breakpoints and corresponding Air Quality Index value
aqi_breakpoints = [
    (0,12.0,50),
    (12.1,35.4,100),
    (35.5,55.4,150),
    (55.5, 150.4, 200), 
    (150.5, 250.4, 300), 
    (250.5, 350.4, 400),
    (350.5, 500.4, 500)
]
```

```python
#function to calculate the aqi for one pollutant
def calculate_aqi (pollutant_name,concentration):
    for low,high,aqi in aqi_breakpoints:
        if low <= concentration <= high:
            return aqi
    return None
```

```python
#calculate overall maximum aqi for a day or row
def calculate_overall_aqi(row):
    aqi_value = []
    pollutant = ['co','no','no2','o3','so2','pm2_5','pm10','nh3']
    for poll in pollutant:
        aqi = calculate_aqi(poll,row[poll])
        if aqi is not None:
            aqi_value.append(aqi)
    return max(aqi_value)
```

```python
#calculate AQI for each row and update the same in new column
Aqi_df['AQI'] = Aqi_df.apply(calculate_overall_aqi,axis=1)
```

```python
Aqi_df.head(10)
```

categorize the AQI based on the value and add the details in new category column

```python
aqi_cat = [
    (0,50,'Good'),
    (51,100,'Moderate'),
    (101,150,'Unhealthy for sensitive people'),
    (151,200,'Unhealthy'),
    (201,300,'Very unhealthy'),
    (301,500,'Hazardous')
]
```

```python
def categorize_aqi(aqi_value):
    for low,high,value in aqi_cat:
        if low <= aqi_value <= high:
            return value
    return None
```

```python
Aqi_df['AQI_Category'] = Aqi_df['AQI'].apply(categorize_aqi)
```

Lets find the hourly average trend of AQI

```python
hourly_avg_aqi = Aqi_df.groupby('hour')['AQI'].mean().reset_index()
hourly_avg_aqi
```

```python
plt.figure(figsize=(30,10))
sns.lineplot(data=hourly_avg_aqi,x='hour',y='AQI')
```

```python
#importing neccessary libraries 
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

Target page = https://alba-watches.co.in/collections/men

```python
# Get the data from website using requests method
request = requests.get('https://alba-watches.co.in/collections/men')
request
```

```python
request.content
```

```python
# from the above requested information we need to get the html data using beatifulsoup method
soup = BeautifulSoup(request.content,'html.parser')
```

```python
soup
```

```python
#Need to fiter the watch model names from the above html code
model = soup.find_all('div',{"class":"grid-product__title"})
```

```python
model
```

```python
model = [i.text for i in model]
```

```python
model
```

```python
price = soup.find_all('div',{"class":"grid-product__price"})
```

```python
#strip the value from span tag
mrp_price = []
for item in price:
    value = item.find('span')
    if value:
        mrp_price.append(value.text.strip())
```

```python
mrp_price
```

```python
#take only the amount value from the mrp string
price = []
for val in range(0,(len(mrp_price))):
    price.append(mrp_price[val][6:])
```

```python
price
```

```python
#Adding all data i
data = {'Model':model,'Price':price}
```

```python
watch_data = pd.DataFrame(data)
```
